```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Basic Graph

```python
x = [1,2,3,4,]
y = [2,4,6,8]
#First Line
plt.plot(x,y,label = 'y = 2x',marker = '.', markersize = 12, linestyle = '--')
#Second Line 

x2 = np.arange(0,4.5,0.5)
print (x2)
plt.plot(x2,x2**2,color='r',label = 'y = x^2')
plt.title("First Graph")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
```

# Bar Graphs

```python
labels = ['A','B','C']
values = [1,4,2]

bars= plt.bar(labels,values)
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')
```
